<h1>Cardiovascular Disease Dataset EDA</h1>

Cardiovascular diseases (CVDs) are a leading cause of mortality worldwide, motivating data-driven approaches that support early identification of high-risk individuals. In this notebook, we perform exploratory data analysis (EDA) on a CVD dataset to summarize cohort characteristics, inspect feature distributions, and examine relationships among clinical and lifestyle variables as a foundation for risk modeling. 

<h2>Importing dataset and features of the data</h2>

```python
import warnings 
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid',{'axes.grid':True, 'grid.color':'gray','grid.alpha':0.7})

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.inspection import permutation_importance

plt.rcParams.update({
    'axes.titlesize':20,
    'axes.titleweight':'bold',
    'axes.labelsize':18,
    'axes.labelweight':'bold',
    'xtick.labelsize':14,
    'ytick.labelsize':14,
    'legend.fontsize':16
})

def style_axes(ax):
    for spine in ax.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(2)
    ax.tick_params(axis='both',which='major',length=8,width=2,colors='black')

print('Import and setup complete')
```

```python
df= pd.read_csv('/kaggle/input/cardiovascular-disease-risk-assessment-dataset/CVD Dataset.csv', encoding='ascii', delimiter=',')

print('Dataset loaded. shape:',df.shape)
df.head(3)
```

```python
#Converting column name to remove extra sppace if needed
df.columns = [col.strip() for col in df.columns]
print(df.columns)
```

```python
#Descriptive statistics for numerical features
df.describe().T
```

```python
#Dataset structure and datatype
df.info()
```

```python
#Checking missing vlaues
df.isnull().sum()
```

<h2>Histogram of age distribution</h2>

```python
plt.figure(figsize=(10,6))

ax=sns.histplot(df["Age"],kde=True,color = "skyblue",edgecolor='black',linewidth=2)
ax.set_title('Age distribution',fontsize=28,fontweight='bold',color='black')
ax.set_xlabel('Age', fontsize=18,fontweight='bold',color='black')
ax.set_ylabel('Frequency', fontsize=18,fontweight='bold',color='black')
style_axes(ax)
plt.tight_layout()
plt.show()
    
```

<h2>Box plot: BMI Across Sex</h2>

Analysing body mass differences across sexes using box plot

```python
plt.figure(figsize=(12,6))
ax=sns.boxplot(data=df, x='Sex', y='BMI', palette='viridis', width=0.6,linewidth=2)
ax.set_title('BMI Across sex', fontsize=20, fontweight='bold',color='black')
ax.set_xlabel('Sex',fontsize=18, fontweight='bold',color='black')
ax.set_ylabel('BMI',fontsize=18, fontweight='bold',color='black')
style_axes(ax)
```

Not much difference is observed between BMI of males and females

<h2>Correlation Heatmap : Numeric features</h2>

```python
numeric_df= df.select_dtypes(include=[np.number])

if numeric_df.shape[1] >=4:
    plt.figure(figsize=(12,10))
    corr= numeric_df.corr()
    sns.heatmap(corr, annot= True, cmap= 'YlGnBu', fmt='.2f',linewidth=2, linecolor='black')
    plt.title('Correlation heatmap of numeric features', fontsize=20,fontweight='bold',color='black')
    plt.xticks(rotation=45,ha='right',fontsize=14,fontweight='bold',color='black')
    plt.yticks(fontsize=14, fontweight='bold', color='black')
    plt.tight_layout()
    plt.show()
```

From the above graph we can figure out the correlation between the different parameters and their intensity of correlation

<h2>Violin plot: CVD risk score by physical activity level</h2>

```python
plt.figure(figsize=(12,10))
ax= sns.violinplot(data=df, x='Physical Activity Level', y='CVD Risk Score',palette='magma', linewidth=2)
ax.set_title('CVD risk score by physical activity level', fontsize=20, fontweight='bold',color='black')
ax.set_xlabel('Physical activity level', fontsize=18,fontweight='bold',color='black')
ax.set_ylabel('CVD risk score', fontsize=18, fontweight='bold',color='black')
style_axes(ax)
plt.tight_layout()
plt.show()
```

<h2>Count plot: Distribution of CVD risk level</h2>

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(data=df, x='CVD Risk Level', palette='coolwarm',edgecolor='black',linewidth=2)
ax.set_title('Distribution of CVD risk level', fontsize=20,fontweight='bold',color='black')
ax.set_xlabel('CVD risk level', fontsize=18,fontweight='bold',color = 'black')
ax.set_ylabel('Count',fontsize=18,fontweight='bold',color='black')
style_axes(ax)

plt.tight_layout()
plt.show()
```

We can see that CVD risk level in out data is mostly high crossing 700+ counts wheres the lowest count is still 200+

<h2>Grouped bar plot: Mean total cholesterol by Diabetes status</h2>

```python
plt.figure(figsize=(12,10))
grouped=df.groupby('Diabetes Status')['Total Cholesterol (mg/dL)'].mean().reset_index()
ax=sns.barplot(data= grouped, x='Diabetes Status', y ='Total Cholesterol (mg/dL)',palette='cubehelix', edgecolor='black',linewidth=2)
ax.set_title('Mean Total Cholesterol by Diabetes Status', fontsize=20, fontweight='bold', color='black')
ax.set_xlabel('Diabetes Status', fontsize=18, fontweight='bold', color='black')
ax.set_ylabel('Mean Total Cholesterol (mg/dL)', fontsize=18, fontweight='bold', color='black')
style_axes(ax)
plt.tight_layout()
plt.show()
```
